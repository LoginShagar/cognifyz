```python
#the top three most  common cuisines in the dataset
cuisine = df['Cuisines'].str.split(',').explode().str.strip().value_counts()
top_cuisine = cuisine.head(3)
print("The top three cuisines are",top_cuisine)
```

```python
#Calculate the percentage of restaurants that serve each of the top cuisines
restaurants_no =df['Restaurant ID'].nunique()
perc=(cuisine/restaurants_no)*100
top= perc.head(3)
print(top)
```

```python
#Identify the city with the highest number of restaurants in the dataset
max_city =df['City'].value_counts().idxmax()
print ("The city with the highest number of restaurants is" ,max_city)
```

```python
#Calculate the average rating for restaurants in each city
avg =df.groupby('City')['Aggregate rating'].mean()
print(avg)
```

```python
#Determine the city with the highest average rating
highest_city = avg.idxmax()
print(highest_city)
```

```python
#Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants
sns.countplot(x='Price range', data=df)
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Price Ranges')
```

```python
# Calculate the percentage of restaurants in each price range category.
price_perc =  df['Price range'].value_counts(normalize=True) * 100
price_perc 
```

```python
# Determine the percentage of restaurants that offer online delivery
perc_deliv =df['Has Online delivery'].str.lower().eq('yes').mean() * 100
perc_deliv
```

```python
#Compare the average ratings of restaurants with and without online delivery
pivot_table = df.pivot_table(values='Aggregate rating', columns='Has Online delivery', aggfunc='mean')
print(pivot_table)
```

```python
# Analyze the distribution of aggregate ratings and determine the most common rating range.
plt.hist(df['Aggregate rating'], bins=10, edgecolor='white')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Aggregate Ratings')
plt.show()
bins = [0, 2, 3, 4, 5]
lab = ['0-2', '2-3', '3-4', '4-5']
df['Rating Range'] = pd.cut(df['Aggregate rating'], bins=bins, labels=lab, include_lowest=True)
rate = df['Rating Range'].value_counts().sort_index()

print("the most common rating range is",rate)
common_rate = rate.idxmax()
print("most common rating range is from",common_rate)
```

```python
#Calculate the average number of votes received by restaurants
avg_votes = df['Votes'].mean()
print("the average number of votes received by restaurantsis ",avg_votes)
```

```python
# Identify the most common combinations of cuisines in the dataset
df['Cuisines'] = df['Cuisines'].fillna('').astype(str)
df['Cuisines'] = df['Cuisines'].apply(lambda x: ', '.join(sorted(x.split(','))))
common_combination = df['Cuisines'].mode().iloc[0]
print(f"most common cuisine combination: {common_combination}")
```

```python
#Determine if certain cuisine combinations tend to have higher ratings
df['Cuisines'] = df['Cuisines'].astype(str).apply(lambda x: ', '.join(sorted(x.split(','))))
avg_rating = df.groupby('Cuisines')['Aggregate rating'].mean()
avg_rating = avg_rating.sort_values(ascending=False)
top = avg_rating.head(10)
top
```

```python
#Plot the locations of restaurants on a map using longitude and latitude coordinates.
plt.figure(figsize=(10, 8))
plt.scatter(df['Longitude'], df['Latitude'], color='green', marker='o', s=20, alpha=0.7)
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
#Identify any patterns or clusters of
def assign_cluster(row):
    if row['Longitude'] > 120 and row['Latitude'] > 14.5:
        return 'Cluster 1'
    elif row['Longitude'] > 120 and row['Latitude'] <= 14.5:
        return 'Cluster 2'
    elif row['Longitude'] <= 120 and row['Latitude'] > 14.5:
        return 'Cluster 3'
    else:
        return 'Cluster 4'
    
df['Cluster'] = df.apply(assign_cluster, axis=1)
colors = {'Cluster 1': 'red', 'Cluster 2': 'blue', 'Cluster 3': 'green', 'Cluster 4': 'purple'}
plt.figure(figsize=(10, 8))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'].apply(lambda x: colors.get(x, 'black')), s=10, alpha=0.5)
plt.title('Restaurant Clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
#Identify if there are any restaurant chains
rest_count = df['Restaurant Name'].value_counts()
chain = rest_count[rest_count > 1]
print("Restaurant chains in the dataset:")
print(chain)
```

```python
# Analyze the ratings and popularity of different restaurant chains
chain_name = chain.index
df_chains = df[df['Restaurant Name'].isin(chain_name)]
chain_agg = df_chains.groupby('Restaurant Name').agg({ 'Aggregate rating': 'mean', 'Votes': 'sum'}).reset_index()
chain_summary_sorted_by_rating = chain_agg.sort_values(by='Aggregate rating', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(chain_summary_sorted_by_rating['Restaurant Name'], chain_summary_sorted_by_rating['Aggregate rating'], color='green')
plt.xlabel('Restaurant Chain')
plt.ylabel('Average Rating')
plt.title('Average Ratings of Restaurant Chains')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()

```

```python
# Calculate the average length of reviews and explore if there is a relationship between review length and rating.
df['Review Length'] = df['Rating text'].apply(len)
average_length = df['Review Length'].mean()
print(f"average review length: {average_length:.2f} characters")
length_by_rating = df.groupby('Aggregate rating')['Review Length'].mean()
print(length_by_rating)

plt.figure(figsize=(8, 5))
length_by_rating.plot(kind='bar', color='green')
plt.xlabel('Rating')
plt.ylabel('Average Review Length')
plt.title('Relationship Between Rating and Average Review Length')
plt.show()
# Calculate correlation 
correlation = df['Review Length'].corr(df['Aggregate rating'])
print(f"the correlation between Review Length and Aggregate Rating: {correlation:.2f}")
```

# wich means that there is no strong relationship between review length and rating

```python
# Analyze the text reviews to identify the most common positive and negative keywords
df['reviews'] = df['Rating text'].apply(lambda x: x.lower().replace('.', '').replace(',', '').split())

positive_rev = ['good', 'very good', 'excellent']
negative_rev = ['not rated', 'average', 'poor']
positive_count= {}
negative_count= {}

for revs in df['reviews']:
    for revs in revs:
        if revs in positive_rev:
            positive_count[revs] = positive_count.get(revs, 0) + 1
        if revs in negative_rev:
            negative_count[revs] = negative_count.get(revs, 0) + 1
            
print("positive reviews:", positive_count)
print("negative reviews:", negative_count)

```

```python
# Identify the restaurants with the highest and lowest number of votes
high_vote = df.loc[df['Votes'].idxmax()]
low_vote = df.loc[df['Votes'].idxmin()]

# Display results
print("Restaurant with highest votes:", high_vote['Restaurant Name'], high_vote['Votes'])
print("Restaurant with lowest votes:", low_vote['Restaurant Name'], low_vote['Votes'])
```

```python
#Analyze if there is a correlation between the number of votes and the rating of a restaurant.
correlation = df['Votes'].corr(df['Aggregate rating'])
print(f"Correlation between number of votes and rating: {correlation:.2f}")
plt.figure(figsize=(6, 2))
sns.scatterplot(data=df, x='Votes', y='Aggregate rating')
sns.regplot(data=df, x='Votes', y='Aggregate rating', scatter=False, color='green')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.title('Scatter Plot of Number of Votes vs. Aggregate Rating')
print(" correlation of 0.31 suggests a moderate tendency for higher-rated restaurants to receive more votes, but it's not a very strong relationship.")
```

```python
#Analyze if there is a relationship between theprice range and the availability of online delivery and table booking.
relation = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].mean()
print(relation)
relation = relation.reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Price range', y='Has Online delivery', data=relation, color='green', label='Online Delivery')
sns.barplot(x='Price range', y='Has Table booking', data=relation, color='red', label='Table Booking')
plt.title('Availability of Online Delivery and Table Booking by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Proportion of Restaurants')
plt.legend(title='Service Availability')
```

# restaurants with higher price ranges (3 and 4) have significantly more restaurants offering table booking compared to lower price ranges.

```python
# Determine if higher-priced restaurants are more likely to offer these services
corr_online_delivery = df['Price range'].corr(df['Has Online delivery'])
corr_table_booking = df['Price range'].corr(df['Has Table booking'])

print(f"Correlation between Price Range and Online Delivery: {corr_online_delivery:.2f}")
print(f"Correlation between Price Range and Table Booking: {corr_table_booking:.2f}")
```

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Price range', y='Has Online delivery', data=df, logistic=True, ci=None, color='blue')
plt.title('Price Range vs. Online Delivery')
plt.xlabel('Price Range')
plt.ylabel('Online Delivery (1 = Yes, 0 = No)')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
sns.regplot(x='Price range', y='Has Table booking', data=df, logistic=True, ci=None, color='red')
plt.title('Price Range vs. Table Booking')
plt.xlabel('Price Range')
plt.ylabel('Table Booking (1 = Yes, 0 = No)')
plt.grid(True)
plt.show()
```
